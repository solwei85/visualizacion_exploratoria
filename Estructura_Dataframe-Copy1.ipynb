```python
#Average Flight Speed: Distance/Airtime. 
#Primero eliminamos las filas vacias de Airtime (distancia no tiene filas vacias)
flights_df.dropna(subset=["AirTime"])
```

```python
#Creamos una columna nueva con esta información.
flights_df["AvFlightSpeed"]= flights_df["Distance"] / flights_df["AirTime"]

#comprobamos
ave_flight_speed= flights_df[["Distance", "AirTime", "AvFlightSpeed"]]
ave_flight_speed.head()
```

```python
#Para la columna de "OnTime", vamos a comprobar si hay diferencia entre arrival time vs CRS/Scheduled arrival time.
flights_df.dropna(subset=["ArrTime"])
```

```python
flights_df["OnTime"]= flights_df["ArrTime"] - flights_df["CRSArrTime"]
On_Time= flights_df[["ArrTime", "CRSArrTime", "OnTime"]]
On_Time.head()
#Si es cero, llegó a tiempo. Si es positivo, llegó tarde.
```
